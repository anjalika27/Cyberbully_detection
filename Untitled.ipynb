#### Encoding categories 

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()

data['cyberbully_type_encoded']=label_encoder.fit_transform(data['cyberbullying_type'])
data[['cyberbullying_type','cyberbully_type_encoded']].value_counts()
```

```python
data.sample(10)
```

### Text Preprocessing

```python
import re
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
ps=PorterStemmer()

corpus=[]

for x in data['tweet_text']:
    tweet=re.sub('[^a-zA-Z]',' ',x)
    tweet=tweet.lower()
    tweet=tweet.split()
    text=[ps.stem(word) for word in tweet if word not in stopwords.words('english')]
    text=' '.join(text)
    corpus.append(text)
#     print(text)
corpus
```

```python
import tensorflow as tf
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=500000,ngram_range=(1,2))
tfidf_corpus=tfidf.fit_transform(corpus)
tfidf.vocabulary_
```

#### Word Cloud

```python
text=' '.join(corpus)

from wordcloud import WordCloud, STOPWORDS
stopwords=set(STOPWORDS)
wordcloud=WordCloud(width=1000,height=500,stopwords=stopwords).generate(text)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

We can see from the wordcloud that there were frequent tweets about:
1. Muslim religion
2. High School 
3. people appearance, etc.

### Model Building

```python
#final independent and dependent features
X=tfidf_corpus
y=data['cyberbully_type_encoded']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=30,test_size=0.2)
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
model_nb=nb.fit(X_train.toarray(),y_train)
y_nb=model_nb.predict(X_test.toarray())
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_nb,y_test))
```

```python
print(classification_report(y_nb,y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
model_knn=knn.fit(X_train,y_train)
y_knn=model_knn.predict(X_test)
```

```python
print(accuracy_score(y_knn,y_test))
```

```python
print(classification_report(y_knn,y_test))
```

```python
from sklearn.svm import SVC
svc=SVC()
model_svc=svc.fit(X_train,y_train)
y_svc=model_svc.predict(X_test)
```

```python
print(accuracy_score(y_svc,y_test))
```

```python
print(classification_report(y_svc,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
model_rfc=rfc.fit(X_train,y_train)
y_rfc=rfc.predict(X_test)
```

```python
print(accuracy_score(y_rfc,y_test))
```

```python
print(classification_report(y_rfc,y_test))
```
